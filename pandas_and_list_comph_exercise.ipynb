```python
#Check the type of the embarked variable. Change its type to category and check again.

df['embarked'].dtypes

df['embarked'] = df['embarked'].astype('category')

df['embarked'].dtypes
```

```python
#Show all the wisdoms of those whose embarked value is C.

df[df['embarked'] == 'C'].head()
```

```python
#Show all the wisdoms of those whose embarked value is not S.
df[df['embarked'] != 'S'].head()
```

```python
#Show all information for passengers who are women and under 30 years old.

df[(df['sex'] == 'female') & (df['age']< 30)].head()
```

```python
#Use Fare to show the information of passengers over 500 or over 70 years old.
df[( df['fare'] > 50 ) | (df['age'] > 70 )].head()
```

```python
#Find the sum of the null values in each variable.
df.isnull().sum()
```

```python
#Remove the who variable from the dataframe.
df = df.drop('who',axis =1 )
df.head()
```

```python
#Fill the empty values in the deck variable with the most repeated value (mode) of the deck variable.
type(df["deck"].mode())
df['deck'].mode()
df["deck"].fillna(df["deck"].mode()[0], inplace=True)
df["deck"].isnull().sum()
```

```python
#Fill in the empty values in the age variable with the median of the age variable.
df['age'].isnull().sum()
df['age'].median()
df['age'].fillna(df['age'].median(),inplace = True)
df['age'].isnull().sum()
```

```python
#Find the sum, count, mean values of the survived variable in the breakdown of pclass and gender variables.

df.groupby(['pclass','sex']).agg({'survived': ['sum','count','mean']})
```

```python
'''
Write a function that will give 1 to those under 30 and 0 to those equal to or above 30. titanic data using the function you wrote
Create a variable named age_flag in the set.
'''
def age (age):
    if age < 30 :
        return 1
    else :
        return 0
df["age_flag"] = df['age'].apply(lambda x : age(x))

df['age_flag'].head(20)
            
            
    
```

```python
df_tip = sns.load_dataset('tips')
df_tip
```

```python
#Find the sum, min, max and average of total_bill values according to the categories (Dinner, Lunch) of the Time variable.
df_tip.groupby('time').agg({"total_bill": ["sum","min","mean","max"]})
```

```python
#Find the sum, min, max and average of total_bill values according to days and time.
df_tip.groupby(['day','time']).agg({"total_bill": ["min","mean","max"]})
```

```python
#Find the sum, min, max and average of the total_bill and type values of the lunch time and female customers according to the day.

df_tip[(df_tip['time'] == 'Lunch')& (df_tip['sex'] == 'Female')].groupby('day').agg({'total_bill' : ['min','max','mean'],
                                                                                        'tip' : ['min','max','mean']})

```

```python
#What is the average of orders with size less than 3 and total_bill greater than 10?

df_tip.loc[ (df_tip['size'] < 3) & (df_tip['total_bill' ]> 10 ), 'total_bill'].mean()
```

```python
#Create a new variable named total_bill_tip_sum. Let it give the total bill and tip paid by each customer.

df_tip['total_bill_tip_sum'] = df_tip['total_bill'] + df_tip['tip']
```

```python
df_ =df_tip['total_bill_tip_sum'].sort_values(ascending = False)[:30]
```
